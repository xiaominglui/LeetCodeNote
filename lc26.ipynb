# 删除排序数组中的重复项

```python
from typing import List
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) > 0:
            last = -10 ** 4
            result = 0
            for i in range(len(nums)):
                print(f'nums[{i}]={nums[i]}===last={last}===result={result}')
                if i == 0:
                    last = nums[0]
                    result = 1
                else:
                    if nums[i] <= last:
                        print(f'==compute first non-duplicated')
                        for d in range(1, len(nums)-i):
                            print(f'======d={d}===nums[{i+d}]={nums[i+d]}')
                            if nums[i+d] > last:
                                print(f'======first non-duplicated found')
                                nums[i] = nums[i+d]
                                last = nums[i+d]
                                result = i + 1
                                print(f'======nums[{i}] and last changed to {nums[i+d]}')
                                break
                            else:
                                print(f'======search continue')
                                continue
                        print(f'======no more, end')
                    else:
                        last = nums[i]
                        result = i + 1
                        print(f'==NOT duplicated, continue===result={result}===last={last}')
            return result
        else:
            return 0
        
```

```python
n = [-5, -4, -4, -3, -2, -2, -2, 0, 0, 1, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2]
```

```python
s = Solution()
s.removeDuplicates(n)
```

方案超出时间限制

看完标准答案后了解到，上面方案的解题思路是正确的，但在过程处理中的方法不好，时间复杂度偏高O(n^2)。后面合适时候可以分析下时间复杂度

```python
from typing import List
class SolutionPro:
    def removeDuplicatesPro(self, nums: List[int]) -> int:
        print(f'=====')
        if len(nums) <= 0:
            return 0
    
        fast = 1
        slow = 1
    
        while (fast < len(nums)):
            print(f'=====before===fast={fast}---slow={slow}')
            if nums[fast] != nums[fast - 1]:
                nums[slow] = nums[fast]
                slow = slow + 1
            print(f'=====after===fast={fast}---slow={slow}')
            fast = fast + 1
            
        return slow
```

```python
n = [0, 1, 2, 2, 3, 4]
```

```python
s = SolutionPro()
s.removeDuplicatesPro(n)
```
